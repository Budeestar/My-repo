```python
"""This program implements gradient descent on a provided dataset."""
import numpy as np
import random
import matplotlib.pyplot as plt
data=np.loadtxt('data.csv',delimiter=',')#read data from file
x=data[:,0].reshape(len(x),1)#extract first column of data(represents DISTANCE CYCLED)
y=data[:,1].reshape(len(y),1)#extract second column of data(represents CALORIES BURNED)
```

```python
'Our predicted value of y will be mx+b'
m=random.random() #random value assigned to m
b=random.random() #random value assigned to b
print(m,b)
```

```python
'''adjust number of iterations and learning rate from here!'''
num_iters=1000000 #number of iterations of Gradient descent
learning_rate=0.000001 #Learning rate of Gradient descent
J=np.zeros((num_iters,1))
```

```python
'''Here we run gradient descent and also calculate mean squared error(cost) of data.'''
for i in range(num_iters):
    J[i]=(1/len(x))*np.sum((m*x+b)-y)**2
    m-=learning_rate*2*np.sum(((m*x+b)-y)*x)
    b-=learning_rate*2*np.sum((m*x+b)-y)
    if(i%(num_iters/10)==0):
        print('Cost-',J[i])
```

```python
'''Plot of cost vs number of iterations.
Also the final plot of the fit on our dataset'''
plt.plot(J,np.arange(num_iters)) #plotting cost vs number of iterations
plt.axis([0,num_iters,-1,5])
plt.xlabel('Number of iterations')
plt.ylabel('Cost')
plt.title('Cost vs Number of Iterations')
plt.show()
plt.scatter(x,y,marker='.',color='green') #plotting our dataset
plt.xlabel('Distance cycled')
plt.ylabel('Calories burned')
plt.title('Distance cycled VS Calories burned')
plt.plot(x,(m*x+b)) #plotting the line of best fit on our dataset
plt.grid()
plt.show()
```
